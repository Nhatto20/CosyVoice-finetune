```python
import os
import shutil
import demucs.separate
import shlex

# ====== Cấu hình ======
input_folder = r"C:\Users\japan\Downloads\mcv-scripted-vi-v23.0\cv-corpus-23.0-2025-09-05\vi\outputs"    # Folder chứa file .wav gốc
output_folder = r"C:\Users\japan\Downloads\mcv-scripted-vi-v23.0\cv-corpus-23.0-2025-09-05\vi\denoised"  # Folder lưu kết quả chỉ còn vocals

model_name = "htdemucs"           # Hoặc htdemucs_ft cho fine-tuned, mặc định dùng htdemucs

# Tạo folder output nếu chưa có
os.makedirs(output_folder, exist_ok=True)

# Lấy danh sách file .wav trong folder
wav_files = [f for f in os.listdir(input_folder) if f.lower().endswith(".wav")]

for wav_file in wav_files:
    input_path = os.path.join(input_folder, wav_file)
    print(f"Processing {wav_file}...")

    # Chạy Demucs để tách vocals
    demucs.separate.main(shlex.split(f"--two-stems vocals -n {model_name} \"{input_path}\""))

    # Kết quả sẽ nằm trong: separated/htdemucs/TRACK_NAME/vocals.wav
    separated_path = os.path.join("separated", model_name, os.path.splitext(wav_file)[0], "vocals.wav")
    
    if os.path.exists(separated_path):
        # Copy file vocals vào folder output, giữ tên gốc
        shutil.copy(separated_path, os.path.join(output_folder, wav_file))
        print(f"Saved vocals to {os.path.join(output_folder, wav_file)}")
    else:
        print(f"Error: {separated_path} not found!")

print("All files processed.")
```
